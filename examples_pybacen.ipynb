<h3>IPCA - Índice Nacional de Preços ao Consumidor Amplo</h3>

```python
ipca = ts.read_time_series(bacen_code=433, start='2010-01-01', end='2021-10-31')
ipca.head()
```

<h3>IGP-M - Índice Geral de Preços do Mercado</h3>

```python
igpm = ts.read_time_series(bacen_code=189, start='2010-01-01', end='2021-10-31')
igpm.head()
```

```python
ts.line_plot(dfs = [[ipca, 'IPCA', 'green'], [igpm, 'IGP-M', 'blue']],
              xtitle = 'Date',
              ytitle = 'Índice',
              title = 'IPCA x IGP-M',
              template="ggplot2"
)
```

<h3>CDI - Certificado de Depósito Interbancário</h3>

```python
cdi = ts.read_time_series(bacen_code=4389, start='2010-01-01', end='2021-10-31')
cdi.head()
```

<h3>Selic - Sistema Especial de Liquidação e de Custódia</h3>

```python
selic = ts.read_time_series(bacen_code=432, start='2010-01-01', end='2021-10-31')
selic.head()
```

```python
ts.line_plot(dfs = [[cdi, 'CDI', 'red'],
                    [selic, 'Selic', 'blue']

                   ],
              xtitle = 'Date',
              ytitle = 'Values',
              title = 'CDI X Selic', 
              template='ggplot2'
)
```

<h2>Dados Yahoo Finance</h2>


```python
from pybacen.yahoo_finance import stocks
```

```python
stk = stocks.read_stock_quote(stock_code= ['ITUB4.SA', 'BBDC4.SA', 'SANB11.SA'], start='2021-01-01', pivot_table=False)


stk.head(5)
```

```python
stocks.candlestick(stk, mov_avg={7: "yellow", 25: "blue"})
```

```python
stocks.describe(stk, 'adjclose')
```

```python
stocks.boxplot(stk, y='adjclose', x='stock_code', color='stock_code', template='plotly_dark')
```
